```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("Apocolypse Food Prep.xlsx")
df.head()
```

```python
# df.Date = pd.to_datetime(pd['Date'])
import random

df.loc[:,'Selling_Price'] = df.Price.map(lambda x : round((x - (x/random.randint(2,3))),2))
df
```

```python
sns.relplot(data=df, x = 'Store', y = 'Selling_Price',kind='line', hue='Product',col='Product')
```

```python
sns.relplot(data=df, x = 'store', y='Price', kind='line', hue='Product',col='Store')
```
