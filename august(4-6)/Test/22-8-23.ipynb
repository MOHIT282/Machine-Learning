Q6.What are various ways to access elements in a series ?

Ans:- 
Let 'a' be a pandas Series

To access multiple or single elements from a Series :-

1  a.loc[labels]
2  a.iloc[indexes or positions]
3  a[]

To access single elements only from a Series :-

1  a.at[label]
2  a.iat[index]

7 (a). Find the names of the person whose age are less than 30 and their number

```python
details = df[df['Age'] < 30]
details.drop(columns=['Age'])
```

7 (b). Find the age of the person who has highest score

```python
df[df['Score'] == np.max(df['Score'])].Age[0]
```

7 (c). Find the sum of ages of those persons who have scored second highest and second lowest score

```python
last_index = (df.count() - 1)[0]
# df.iloc[ second_last ]
df.iloc[1].Age + df.iloc[last_index - 1].Age
```

7 (d). Find the sum of points of persons whose name starts with ‘G’

```python
df
```

```python
def NameG(x):
    
    if(x[0].lower() == 'g'):
        return True
    else:
        return False
    
df[df.Driver.apply(lambda x : NameG(x))].Score.sum()
```

7 (e). Find the name of persons whose age is lie between 33 to 40

```python
df[df.Age.between(33,40)].Driver.values
```

7 (f). Sort the drivers list by their age

```python
df.sort_values('Age', ignore_index=True)
```
