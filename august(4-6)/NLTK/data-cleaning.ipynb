- remove punctuations and escape characters

```python

def remove(text):
    esc_ch = ['\n','\t','\r']
    punc = string.punctuation
    text = demoji.replace_with_desc(text)
    
    for ch in text:
        if ch in punc or ch in esc_ch:
            text = text.replace(ch,'')
    
    
    # replace emojis with their meanings
    
    return text
```
