```python
# import spacy for doing lemmatization
import spacy
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv('cleaned_tweets.csv')
df.head(10)
```

```python
df.text.unique()
```

```python
def lemmatizer(text):
    str = ""
    doc = nlp(text)
    for i in doc:
        str += " " + i.lemma_
    
    return str
```

```python
df.text = df.text.astype('str')
df.loc[:,'Lemmatization'] = df.text.apply(lemmatizer)
```
