- handle missing data in gender Column

```python
df[df.Sex.isna()]
```

```python
gen_index = df[df.Sex.isna()].index
gen_index
```

```python
def gender(title):
    
    if(title == 'Mr'):
        return 'male'
    
    elif(title == 'Miss'):
        return 'female'
    
    elif(title == 'Mrs'):
        return 'female'
    
    elif(title == 'Master'):
        return 'male'
    
```

```python
df.loc[gen_index,'Sex'] = df.loc[gen_index,:].Title.apply(gender)
```

```python
df.Sex.isna().any()
```
