```python
nan_index = df[df.worldPercentage.isna()].index
nan_index
```

```python
df.worldPercentage.unique()
```

```python
df['worldPercentage'] = round(df['worldPercentage'], 2)
df.worldPercentage.unique()
```

```python
world_population = df.pop2023.sum()
```

```python
df.loc[pop_nan,'worldPercentage'] = df.loc[pop_nan,'pop2023'].apply(lambda x : (x/world_population)*100)
```
