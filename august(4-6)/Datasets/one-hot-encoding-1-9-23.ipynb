- we can also use get_dummies functions of pandas library to do the whole process in just one line

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('titanic1.csv')
df
```

```python
gender_index = df[df.Sex.isna()].index

df.loc[gender_index,'Sex'] = df.Sex.mode()[0]
```

```python
enc = OneHotEncoder(handle_unknown='ignore', drop='first')

enc.fit_transform(df.loc[:,['Sex']]).toarray()
```

```python
enc.categories_
```

```python
enc.get_feature_names_out()
```

```python
df.loc[:,enc.get_feature_names_out()] = enc.fit_transform(df.loc[:,['Sex']]).toarray()
```

```python
df
```

```python
pd.get_dummies(df.Sex)
```

```python
enc2 = OneHotEncoder(handle_unknown='ignore',drop='first')
enc2.fit_transform(df.loc[:,['Embarked']]).toarray()
```

```python
df.loc[:,enc2.get_feature_names_out()] = enc2.fit_transform(df.loc[:,['Embarked']]).toarray()
df
```
