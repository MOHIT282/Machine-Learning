```python
import pandas as pd
import random
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Define the ordinal categories
ordinal_categories = ['Low', 'Higher', 'High']
```

```python
# Generate random data with ordinal categories
data = []
num_rows = random.randint(100, 200)  # Generate a random number of rows
```

```python
for _ in range(num_rows):
    category = random.choice(ordinal_categories)  # Randomly select a category
    data.append([category])
```

```python
# Create a DataFrame
df = pd.DataFrame(data, columns=['income'])
```

```python
df
```

```python
encoder = OrdinalEncoder()
```

```python
encoder.fit_transform(df.loc[:,['income']])
```

```python
encoder.categories_
```

```python
encoder.get_feature_names_out()
```

```python
df.loc[:,encoder.get_feature_names_out()] = encoder.fit_transform(df.loc[:,['income']])
```
